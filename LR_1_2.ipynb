```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

x = 12.1
c = np.arange(-10, 1.5, 0.5)
print("Все значения c:", c)
l = (abs(2*x - c) ** 0.6) + 0.567
print("Массив значений функций на отрезке", l)
l_max = np.max(l)
print("Наибольшее значение функции на отрезке", l_max)
l_min = np.min(l)
print("Наименьшее значение функции на отрезке", l_min)
l_kol = np.size(l)
print("Количество значений функций в массиве", l_kol)
l_aver = np.sum(l) / l_kol
print("Среднее значение функции на отрезке", l_aver)
l_sorted = np.sort(l)
print("Отсортированный массив", l_sorted)
line = np.full_like(c, l_aver)
plt.plot(c, l)
plt.plot(c, line)
plt.ylabel('C ось')
plt.xlabel('L ось')
plt.show()
```
